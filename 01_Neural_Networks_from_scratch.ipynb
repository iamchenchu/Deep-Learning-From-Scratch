### Neural Network from Scratch

Now let's build the simple neuron with 3 inputs 

### Neuron with 3 inputs




```python
inputs = [2, 4, 6]
weights = [0.5, 0.9, 0.3]
b = 0.6

output = inputs[0] * weights[0] + inputs[1] * weights[1] + inputs[2]* weights[2] + b

print(output)
```

### Neuron with 4 inputs


```python
inputs = [3, 4, 8, 19]
weights = [0.8, 0.4, 0.2, 0.1]
bias = 0.5

output = inputs[0] * weights[0] + inputs[1] * weights[1] + inputs[2] * weights[2] + inputs[3] * weights[3] + bias

print(output)
```

### Layer of Neurons



```python
inputs = [ 1, 2, 3, 2.5]

weights = [[0.2, 0.8, -0.5, 1],
          [0.2, 0.4, -0.1, 0.6],
          [0.9, 0.32, 0.54, 0.12]]

weights1 = weights[0] # List of weights for neuron 1 : w11, 12, w13, w14
weights2 = weights[1] # List of weights for neuron 2 : w21, 22, w23, w24
weights3 = weights[2] # List of weights for neuron 3 : w31, 32, w33, w34

bias1 = 2
bias2 = 5
bias3 = 9

outputs =[
    inputs[0]*weights1[0] + inputs[1]*weights1[1] + inputs[2]*weights1[2] + inputs[3]*weights1[3] + bias1, 
    inputs[0]*weights2[0] + inputs[1]*weights2[1] + inputs[2]*weights2[2] + inputs[3]*weights2[3] + bias2,
    inputs[0]*weights3[0] + inputs[1]*weights3[1] + inputs[2]*weights3[2] + inputs[3]*weights3[3] + bias3
]

print(outputs)
```

```python
## Now make the above cell using the loop 

inputs = [ 1, 2, 3, 2.5]

weights = [[0.2, 0.8, -0.5, 1],
          [0.2, 0.4, -0.1, 0.6],
          [0.9, 0.32, 0.54, 0.12]]

#List of biases
biases = [2, 5, 9]

# Outputs of current layers
layer_outputs = []

for neuron_weights, neuron_bias in zip(weights, biases):
    
    neuron_output = 0

    for n_input, weights in zip(inputs, neuron_weights):

        neuron_output += n_input * weights

    neuron_output += neuron_bias

    layer_outputs.append(neuron_output)


print(layer_outputs)

        
```

### USING Numpy 

The above code can be simplified using the numpy, we just need to understand the dot product of the vectors and matrices 

dot product of vectors is always a scalar 

`np.dot(a, b) == np.(b, a)` # this is possible when a and b both are vectors 

If incase if any of the a or b is a matrix then `np.dot(a, b) != np.dot(b, a)`


### Neuron with 3 inputs



```python
import numpy as np

inputs = [2, 4, 6]
weights = [0.5, 0.9, 0.3]
b = 0.6

output = np.dot(inputs, weights) + b

print(output)
```

### Neuron with 4 inputs




```python
import numpy as np

inputs = [3, 4, 8, 19]
weights = [0.8, 0.4, 0.2, 0.1]
bias = 0.5

output = np.dot(inputs, weights) + bias

print(output)
```

## Layer of neurons



```python
import numpy as np 

inputs = [ 1, 2, 3, 2.5]

weights = [[0.2, 0.8, -0.5, 1],
          [0.2, 0.4, -0.1, 0.6],
          [0.9, 0.32, 0.54, 0.12]]

#List of biases
biases = [2, 5, 9]


layer_outputs = np.dot(weights, inputs) + biases

print(layer_outputs)
```
