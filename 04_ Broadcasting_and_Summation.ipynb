### Broadcasting and array summation in Python



```python
import numpy as np

A = [[1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]]

print(np.sum(A)) #It adds all the elements in the array and give us the total sum of the elements
```

```python
axis_0_sum = np.sum(A, axis=0)

print(axis_0_sum)

print("The dimesion is :", axis_0_sum.ndim)
```

```python
axis_1_sum = np.sum(A, axis=1)

print(axis_1_sum)

print("The dimesion is :", axis_1_sum.ndim)
```

```python
axis_0_sum = np.sum(A, axis=0, keepdims=True)

print(axis_0_sum)

print("The dimesion is :", axis_0_sum.ndim)
```
